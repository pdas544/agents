Commercial Application:

First ask the LLM to pick a business area that might be worth exploring for an Agentic AI opportunity. <br>
Then ask the LLM to present a pain-point in that industry - something challenging that might be ripe for an Agentic solution. <br>
Finally have 3 third LLM call propose the Agentic AI solution.

```python
from dotenv import load_dotenv
load_dotenv(override=True)

import os
openai_api_key = os.getenv('OPENAI_API_KEY')

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set - please head to the troubleshooting guide in the setup folder")
    
from openai import OpenAI

openai = OpenAI()
```

```python
# Pick a business area that might be worth exploring for an Agentic opportunity

question1 = [{"role": "user", "content": "Pick a business area that might be worth exploring for an Agentic opportunity"}]
response = openai.chat.completions.create(model="gpt-5.4-nano", messages=question1)
business_area = response.choices[0].message.content
print(business_area)
```

```python
# Present a pain-point in the industry - something challenging that might be ripe for an Agentic solution

question2 = [{"role": "user", "content": f"Given this business area, come up with a pain point in this business are that could be sold: {business_area}"}]
response = openai.chat.completions.create(model="gpt-5.4-nano", messages=question2)
pain_point = response.choices[0].message.content
print(pain_point)
```

```python
# Propose the Agentic AI solution to the pain-point

question3 = [{"role": "user", "content": f"Given this business area and its pain point, propose an agentic ai solution : {business_area} {pain_point}"}]
response = openai.chat.completions.create(model="gpt-5.4-nano", messages=question3)
solution = response.choices[0].message.content

print(solution)
```
