```python
import os
from typing import override
from dotenv import load_dotenv
from scraper import fetch_website_contents
from IPython.display import Markdown, display
from huggingface_hub import InferenceClient

load_dotenv(override=True)
hf_token = os.getenv("HF_TOKEN")
client = InferenceClient(api_key=os.getenv('HF_TOKEN'))

if not hf_token:
    print("No HF_TOKEN was found - please add it to your .env file!")
elif not hf_token.startswith("hf_"):
    print("A token was found, but it doesn't start with hf_; please check you're using the right key")
else:
    print("HF token found and looks good!")
```

```python
def messages_for(website, system_prompt, user_prompt):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + website}
    ]
```

```python
url = "https://www.ausbildung.de/"
system_prompt = """
You are a translator who specializes in English language. You help people who are Ausbildung seeker in Germany. 
You should translate the German job postings to English. 
Do not wrap the markdown in a code block - respond just with the markdown. Make your response user friendly.
"""
user_prompt = """
    Here is the website. Translate it.
"""
```

```python
def initializeAIModel(messages): 
    response = client.chat.completions.create(
        model="meta-llama/Meta-Llama-3-8B-Instruct",
        messages = messages
    )   
    return response
```
